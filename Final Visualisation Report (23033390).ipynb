```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
data = pd.read_csv('/content/heart_statlog_cleveland_hungary_final.csv')

# Bar Chart: Distribution of Chest Pain Types
plt.figure(figsize=(10, 6))
sns.countplot(x='chest pain type', data=data)
plt.title('Distribution of Chest Pain Types')
plt.xlabel('Chest Pain Type')
plt.ylabel('Count')
plt.show()

# Pie Chart: Proportion of Heart Disease in Patients
heart_disease_counts = data['target'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(heart_disease_counts, labels=['No Heart Disease', 'Heart Disease'], autopct='%1.1f%%', startangle=90)
plt.title('Proportion of Heart Disease in Patients')
plt.show()

# Scatter Chart with Trendline: Cholesterol vs. Max Heart Rate
plt.figure(figsize=(10, 6))
sns.regplot(x='cholesterol', y='max heart rate', data=data, scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.title('Cholesterol vs. Max Heart Rate with Trendline')
plt.xlabel('Cholesterol (mg/dL)')
plt.ylabel('Max Heart Rate')
plt.show()

# Statistical summary for selected columns
summary_stats = data[['age', 'cholesterol', 'max heart rate']].describe()
correlations = data[['cholesterol', 'max heart rate']].corr()

print("Statistical Summary:")
print(summary_stats)
print("\nCorrelations:")
print(correlations)
```
